```python
num_img = 5

for channel_num in range(9):

    result = np.zeros((num_time, len(AoI), 3, 2))
    
    print(channel_num)

    for idx_dir in tqdm(range(num_time)): 
        directory = './'+source_dir+'/t'+str(idx_dir+1)+'/c' + str(channel_num) +'/'
        
        
        temp = np.zeros((num_img, len(AoI), 3))

        for (idx_sample, filename) in enumerate(sorted(os.listdir(directory))):
            path = os.path.join(directory, filename)

            rgb = mpimg.imread(path)

            for (idx_AOI, (i,j)) in enumerate(AoI):
                #extract mean pixel value in r,g,b channel

                for (idx_pixel, mean_pixel) in enumerate(np.mean(rgb[i:i+area, j:j+area], axis = (0,1))):
                    temp[idx_sample, idx_AOI, idx_pixel] = mean_pixel

        result[idx_dir, :, :, 0] = np.mean(temp, axis = 0)
        result[idx_dir, :, :, 1] = np.var(temp, axis = 0)
    
    # Save the array to an .npy file
    # Check if the directory already exists
    print('./cache/'+source_dir)
    if not os.path.exists('./cache/'+source_dir):
        # Create the directory
        os.mkdir('./cache/'+source_dir)
        
    np.save( './cache/'+source_dir+'/output_c'+str(channel_num)+'.npy', result)
    
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,0], rgb[:,1], rgb[:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
```

```python
def moving_average(data, window_size):
    """
    Computes the moving average of the given data using a specified window size.
    
    Args:
        data: The input data as a 1-D NumPy array.
        window_size: The size of the moving average window.
    
    Returns:
        An array containing the moving average of the input data.
    """
    cumsum = np.cumsum(data)
    cumsum[window_size:] = cumsum[window_size:] - cumsum[:-window_size]
    return cumsum[window_size - 1:] / window_size
```
